### Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the Dataset

```python
df = pd.read_csv(r"C:\Users\sathi\Downloads\car_df_EDA.csv")
```

```python
df.head()
```

### Data Cleaning

```python
df.drop(["Unnamed: 0.1","Unnamed: 0"],axis=1,inplace=True)
```

### Structural Errors

```python
pd.set_option("display.max_rows",100)
pd.set_option("display.max_columns",50)
for i in df.columns:
    #print(df[i])
    print(f"{i}",df[i].unique())
    print()
```

```python
df["num_of_doors"].replace("?",np.nan,inplace=True)

df["num_of_doors"].replace("?",np.nan,inplace=True)

df["num_of_doors"].replace(["two","four"],[2,4],inplace=True)

df["num_of_cylinders"].replace(['four','six','five','three','twelve','two','eight'],[4,6,5,3,12,2,8],inplace=True)

df["horsepower"].replace("?",np.nan,inplace=True)

df["peak_rpm"].replace("?",np.nan,inplace=True)

df["price"].replace("?",np.nan,inplace=True)


df.head()
```

### Missing Value Detection

```python
df.isna().sum()
```

```python
df.info()
```

```python
num = df.select_dtypes(exclude="object")
obj = df.select_dtypes(include="object")
```

```python
#filling missing values for numerical features
for i in num.columns:
    #print(num[i])
    df[i].fillna(df[i].median(),inplace=True)
```

```python
df.isna().sum()
```

```python
#filling missing values for categorical features
for i in obj.columns:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
df.isna().sum()
```

### TypeCasting

```python
df.dtypes
```

```python
df["price"] = df["price"].astype("float64")

df["peak_rpm"] = df["peak_rpm"].astype("float64")

df["horsepower"] = df["horsepower"].astype("float64")

df.dtypes
```
