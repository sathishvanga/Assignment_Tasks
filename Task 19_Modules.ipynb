Create a Python script that simulates a simple login system. The script should prompt the user to enter their username and password. If the entered username and password match the predefined credentials, the script should log a successful login event to a log file. If the credentials do not match, it should log a failed login attempt to the log file. Example Output Assuming the predefined credentials are `username: "user123"` and `password: "pass123"`, here's an example output of the script:



Create any 5 of following modules

Do some research while creating them

```python
usernames = []


def signup():
    class EnterValidMobileNumber(Exception):
        pass
    
    global usernames
    
    print(("*"*25),"Create a account!",("*"*25))
    user = input("Create your User Name :: ").lower()

    password = input("Create your password :: ")

    usernames.extend([user,password])

    mobile = int(input("Enter your Mobile Number :: "))
    length = str(mobile)
    
    if len(length) != 10 :
        raise EnterValidMobileNumber("Enter a valid Mobile Number")
        
    print(("*"*15),"Your Account Created Successfully",("*"*15))
    print("\n",("*"*25),"You can Login Now" ,("*"*25))

    
def signin():
    import time as t
    global usernames 
    
    user_name = input("Enter your username :: ")
    password_1 = input("Enter your password :: ")
    if (user_name in usernames) and (password_1 in usernames):
        print(f"Login Successful for {user_name} ",t.strftime("(%m/%d/%Y, %H:%M:%S)"))
    else :
        print(f"Login Failed for {user_name} ",t.strftime("(%m/%d/%Y, %H:%M:%S)"))
        
        
i=1

while i == 1:   
    ask = input("Do you have account Yes/No :: ").capitalize()

    if ask == "Yes":
        signin()
        i=0

    elif ask == "No":
        signup()
        i=1

    else :
        raise ValueError("Enter a valid input!")
        

        
    
    
    
```
