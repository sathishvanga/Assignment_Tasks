#### IQR

```python
arr
```

```python
q1 = np.percentile(arr,25)
q1
```

```python
q2 = np.percentile(arr,50)
q2
```

```python
q3 = np.percentile(arr,75)
q3
```

```python
IQR = q3-q1
IQR
```

```python
lower = q1 - (1.5 * IQR)
upper = q3 + (1.5 * IQR)
```

```python
lower
```

```python
upper
```

```python
arr[arr>upper]
```
