### Statistical Measures on SepalWidth

```python
sepal_width
```

```python
sepal_width.mean()
```

```python
np.median(sepal_width)
```

```python
range = np.max(sepal_width)-np.min(sepal_width)
range
```

```python
q3 = np.quantile(sepal_width,0.75)
q3
```

```python
q1 = np.quantile(sepal_width,0.25)
q1
```

```python
IQR = q3-q1
```

```python
lower = q1 - (1.5*IQR)
upper = q3 + (1.5*IQR)
lower
```

```python
sepal_width[sepal_width<lower]
```

```python
sepal_width[sepal_width>upper]
```

```python
np.var(sepal_width)
```

```python
np.std(sepal_width)
```

### Statistical Measures on SepalLength

```python
sepal_length
```

```python
sepal_length.mean()
```

```python
np.median(sepal_length)
```

```python
range = np.max(sepal_length) - np.min(sepal_length)
range
```

```python
q1 = np.quantile(sepal_length,0.25)
```

```python
q3 = np.quantile(sepal_length,0.75)
q3
```

```python
IQR = q3 - q1
```

```python
lower = q1 - (1.5 * IQR)
upper = q3 - (1.5 * IQR)
upper
```

```python
sepal_length[sepal_length<lower]
```

```python
sepal_length[sepal_length>upper]
```

```python
np.var(sepal_length)
```

```python
np.std(sepal_length)
```
